```python
bios_new = bios.copy()
```

```python
bios_new['first_name'] = bios_new['name'].str.split(' ').str[0]
```

```python
bios_new.query('first_name == "Keith"')
```

```python
bios_new['born_datetime'] = pd.to_datetime(bios_new['born_date'], errors='coerce')
```

```python
bios_new
```

```python
bios_new['born_year'] = bios_new['born_datetime'].dt.year

bios_new[['name','born_year']]
```

```python
bios_new.to_csv('bios_new.csv', index=False)
```

```python
bios_new['height_category'] = bios_new['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall'))
```

```python
bios_new[['name','height_category']]
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] < 80:
        return 'Middleweight'
    else:
        return 'Heavyweight'

bios_new['weight_category'] = bios.apply(categorize_athlete, axis=1)
```

```python
bios_new[['name','height_category','weight_category']]
```
